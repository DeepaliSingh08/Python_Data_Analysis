```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#Drop Unnamed Column

```python
df= df.drop("Unnamed: 0",axis=1)
print(df.head())
```

#Gender Distrubution

```python
plt.figure(figsize=(5,5))
ax=sns.countplot(data=df,x="Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

#From the avobe chart we have analysed that:
#The number of females in tha data is more than the number of males

```python
gb=df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb, annot = True)
plt.title("Relationship B/W Parent's Education and Student's Score")
plt.show()
```

#From the above chart we have concluded that the education of the parents have a good impact on thier scores

```python
gb1=df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb1)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb1, annot = True)
plt.title("Relationship B/W Parent's Marital Status and Student's Score")
plt.show()
```

#From the above chart we have concluded that the Marital Status of the parents have negligible/no major impact on thier scores

#Distribution Of Ethnic Groups

```python
groupA = df.loc[(df['EthnicGroup'] == "group A")].count()
groupB = df.loc[(df['EthnicGroup'] == "group B")].count()
groupC = df.loc[(df['EthnicGroup'] == "group C")].count()
groupD = df.loc[(df['EthnicGroup'] == "group D")].count()
groupE = df.loc[(df['EthnicGroup'] == "group E")].count()

l=["group A","group B","group C","group D","group E"]
mylist=[groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]

print(mylist)

plt.pie(mylist,labels=l, autopct = "%1.2f%%")
plt.title("Distribution of Ethnic Groups")
plt.show()
```
